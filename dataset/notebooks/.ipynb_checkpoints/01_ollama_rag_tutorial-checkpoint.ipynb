## STEP 3 — 임베딩 이해하기

임베딩 = 텍스트를 숫자 배열(벡터)로 변환하는 것  
의미가 비슷한 텍스트는 비슷한 숫자 배열이 됩니다.

```python
from sentence_transformers import SentenceTransformer, util

# bge-m3: 한국어 + 영어 + 일본어 + 100개 언어 지원
# 처음 실행 시 570MB 다운로드됨
print("⏳ bge-m3 모델 로딩 중...")
embedder = SentenceTransformer("BAAI/bge-m3")
print("✅ 모델 로딩 완료")
```

```python
# 텍스트를 벡터로 변환해보기
text = "경복궁 서울 조선시대 궁궐"
vector = embedder.encode(text)

print(f"입력 텍스트: '{text}'")
print(f"벡터 크기: {vector.shape}")
print(f"벡터 앞 10개 숫자: {vector[:10].round(4)}")
print("\n→ 이 숫자 배열이 '경복궁'의 의미를 표현합니다")
```

```python
# 의미가 비슷한 문장끼리 유사도가 높게 나오는지 확인
sentences = [
    "경복궁",                     # 기준
    "조선시대 궁궐 서울",           # 비슷한 의미
    "Gyeongbokgung Palace",       # 영어 (같은 장소)
    "景福宮",                      # 일본어/한자 (같은 장소)
    "삼겹살 맛집 홍대",              # 완전히 다른 주제
    "자전거 도로 한강",              # 다른 주제
]

base = embedder.encode(sentences[0], normalize_embeddings=True)
others = embedder.encode(sentences[1:], normalize_embeddings=True)

print(f"기준 문장: '{sentences[0]}'\n")
for sent, vec in zip(sentences[1:], others):
    similarity = util.cos_sim(base, vec).item()
    bar = "█" * int(similarity * 20)
    print(f"  {similarity:.3f} {bar}  '{sent}'")
```

```python
# 다국어 동일 의미 테스트 — 핵심!
# 한/영/일이 같은 의미면 유사도가 높아야 함
multilingual = [
    ("한국어", "BTS 뮤직비디오 촬영지 서울"),
    ("English", "BTS music video filming location Seoul"),
    ("日本語", "BTSのミュージックビデオ撮影地ソウル"),
]

vecs = embedder.encode([t for _, t in multilingual], normalize_embeddings=True)

print("=== 다국어 유사도 (같은 의미인지 확인) ===")
for i, (lang1, text1) in enumerate(multilingual):
    for j, (lang2, text2) in enumerate(multilingual):
        if i < j:
            sim = util.cos_sim(vecs[i], vecs[j]).item()
            print(f"  {lang1} ↔ {lang2}: {sim:.3f}")

print("\n→ 0.7 이상이면 같은 의미로 인식하는 것")
```

## STEP 5 — 벡터 검색 테스트

이제 질문을 넣으면 관련 POI를 찾아오는지 확인합니다.

```python
def search_poi(query: str, top_k: int = 3) -> list[dict]:
    """질문과 유사한 POI를 벡터DB에서 검색"""
    # 질문을 벡터로 변환
    query_vec = embedder.encode(query, normalize_embeddings=True).tolist()

    # ChromaDB 검색
    results = collection.query(
        query_embeddings=[query_vec],
        n_results=top_k,
        include=["documents", "metadatas", "distances"]
    )

    # 결과 정리
    pois = []
    for meta, dist in zip(results["metadatas"][0], results["distances"][0]):
        pois.append({
            **meta,
            "similarity": round(1 - dist, 3)  # 거리 → 유사도
        })
    return pois

def print_search_result(query: str, results: list[dict]):
    print(f"🔍 질문: '{query}'\n")
    for i, r in enumerate(results, 1):
        bar = "█" * int(r["similarity"] * 10)
        print(f"  {i}. {r['name']} ({r['sido']}, {r['category']})")
        print(f"     유사도: {r['similarity']} {bar}")
        print(f"     주소: {r['address']}")
        print(f"     이미지: {r['image_url']}")
        print()

print("✅ 검색 함수 준비 완료")
```

```python
# 한국어 검색 테스트
results = search_poi("BTS 팬이 꼭 가야 할 성지")
print_search_result("BTS 팬이 꼭 가야 할 성지", results)
```

```python
# 영어 검색 테스트 — 한국어 POI도 찾히는지 확인
results = search_poi("traditional Korean food market")
print_search_result("traditional Korean food market", results)
```

```python
# 일본어 검색 테스트
results = search_poi("韓国の伝統文化を体験できる場所")
print_search_result("韓国の伝統文化を体験できる場所", results)
```

```python
# 자연 관련 검색
results = search_poi("자전거 타기 좋은 곳")
print_search_result("자전거 타기 좋은 곳", results)
```

## STEP 6 — RAG 완성: 검색 결과를 LLM에 전달

이제 검색한 POI를 Ollama LLM에 전달해서 자연스러운 추천 답변을 만듭니다.  
이것이 RAG (Retrieval-Augmented Generation) 의 핵심입니다.

```python
def kride_recommend(user_query: str, lang: str = "ko", top_k: int = 3) -> dict:
    """
    RAG 추천 함수
    1. 벡터DB에서 관련 POI 검색
    2. 검색 결과를 LLM 프롬프트에 포함
    3. LLM이 자연어 답변 생성
    """

    # Step 1: POI 검색
    pois = search_poi(user_query, top_k=top_k)

    # Step 2: 검색 결과를 읽기 쉬운 텍스트로 변환
    context_lines = []
    for i, p in enumerate(pois, 1):
        context_lines.append(
            f"{i}. {p['name']} ({p['sido']}, {p['category']}) — {p['address']}"
        )
    context = "\n".join(context_lines)

    # Step 3: 언어별 프롬프트
    prompts = {
        "ko": f"""당신은 한국 여행 전문 AI 가이드입니다.
아래 검색된 장소만 참고해서 답하세요. 목록에 없는 장소는 절대 만들어내지 마세요.

[검색된 장소]
{context}

[사용자 질문]
{user_query}

친절하고 간결하게 3~5문장으로 답해주세요.""",

        "en": f"""You are a Korean travel AI guide.
Answer ONLY based on the search results below. Never invent places not listed.

[Search Results]
{context}

[User Question]
{user_query}

Answer kindly in 3-5 sentences.""",

        "ja": f"""あなたは韓国旅行の専門AIガイドです。
以下の検索結果だけを参考に答えてください。リストにない場所は絶対に作り出さないでください。

[検索結果]
{context}

[ユーザーの質問]
{user_query}

3〜5文で親切に答えてください。"""
    }

    # Step 4: Ollama LLM 호출
    response = ollama.chat(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompts.get(lang, prompts["ko"])}]
    )

    return {
        "answer": response["message"]["content"],
        "pois": pois,
        "query": user_query,
        "lang": lang
    }


def print_recommendation(result: dict):
    print("=" * 50)
    print(f"질문: {result['query']}  [{result['lang']}]")
    print("=" * 50)

    print("\n💬 AI 답변:")
    print(result["answer"])

    print("\n📍 검색된 장소 + 이미지:")
    for poi in result["pois"]:
        print(f"  - {poi['name']} ({poi['sido']})")
        print(f"    이미지: {poi['image_url']}")


print("✅ RAG 함수 준비 완료")
```

```python
# 한국어 RAG 추천 테스트
result = kride_recommend(
    user_query="K-Pop 팬인데 서울에서 꼭 가야 할 곳 알려줘",
    lang="ko"
)
print_recommendation(result)
```

```python
# 영어 RAG 추천 테스트
result = kride_recommend(
    user_query="I'm a BTS fan. Where should I visit in Seoul?",
    lang="en"
)
print_recommendation(result)
```

```python
# 일본어 RAG 추천 테스트
result = kride_recommend(
    user_query="韓国の伝統文化を感じられる場所に行きたいです",
    lang="ja"
)
print_recommendation(result)
```

## STEP 7 — 실제 K-Ride DB 연결 (PostgreSQL → ChromaDB)

샘플 10개가 잘 작동하면 실제 DB에서 POI를 가져와 대규모로 인덱싱합니다.  
⚠️ `.env` 파일에 `DATABASE_URL`이 설정되어 있어야 합니다.

```python
import os
import psycopg2
from dotenv import load_dotenv

load_dotenv(dotenv_path="../.env")

# DB 연결 확인
try:
    conn = psycopg2.connect(os.getenv("DATABASE_URL"))
    cur = conn.cursor()
    cur.execute("SELECT COUNT(*) FROM poi")
    count = cur.fetchone()[0]
    print(f"✅ DB 연결 성공! POI 테이블: {count:,}개")
    cur.close()
    conn.close()
except Exception as e:
    print(f"❌ DB 연결 실패: {e}")
    print("   .env 파일의 DATABASE_URL을 확인하세요")
```

```python
# image_url 컬럼 추가 (없으면 추가, 이미 있으면 무시)
try:
    conn = psycopg2.connect(os.getenv("DATABASE_URL"))
    cur = conn.cursor()
    cur.execute("""
        ALTER TABLE poi
        ADD COLUMN IF NOT EXISTS image_url VARCHAR(500),
        ADD COLUMN IF NOT EXISTS image_url_thumb VARCHAR(500);
    """)
    conn.commit()
    print("✅ image_url 컬럼 준비 완료")
    cur.close()
    conn.close()
except Exception as e:
    print(f"❌ 에러: {e}")
```

```python
def index_poi_from_db(limit: int = 500, batch_size: int = 64):
    """
    실제 DB의 POI를 ChromaDB에 인덱싱
    limit: 처음 테스트는 500개, 전체는 None
    batch_size: 한 번에 임베딩할 개수 (메모리 절약)
    """
    conn = psycopg2.connect(os.getenv("DATABASE_URL"))
    cur = conn.cursor()

    # 기존 실제 컬렉션 초기화
    try:
        client.delete_collection("kride_poi")
    except:
        pass
    real_collection = client.create_collection(
        name="kride_poi",
        metadata={"hnsw:space": "cosine"}
    )

    # DB에서 POI 조회
    limit_clause = f"LIMIT {limit}" if limit else ""
    cur.execute(f"""
        SELECT id, name, category, sub_category, sido, address,
               COALESCE(image_url, '') as image_url
        FROM poi
        ORDER BY id
        {limit_clause}
    """)
    rows = cur.fetchall()
    print(f"DB에서 {len(rows)}개 POI 조회 완료")

    # 배치 단위로 임베딩 + 저장
    total = 0
    for start in range(0, len(rows), batch_size):
        batch = rows[start:start + batch_size]

        texts, ids, metadatas = [], [], []
        for row in batch:
            id_, name, cat, subcat, sido, addr, img_url = row
            text = f"{name} {cat} {subcat or ''} {sido or ''} {addr or ''}"
            texts.append(text)
            ids.append(str(id_))
            metadatas.append({
                "name": name or "",
                "category": cat or "",
                "sido": sido or "",
                "address": addr or "",
                "image_url": img_url or ""
            })

        embeddings = embedder.encode(texts, normalize_embeddings=True).tolist()
        real_collection.upsert(
            documents=texts,
            embeddings=embeddings,
            metadatas=metadatas,
            ids=ids
        )
        total += len(batch)
        print(f"  진행: {total}/{len(rows)} ({total/len(rows)*100:.0f}%)")

    print(f"\n✅ 인덱싱 완료: {real_collection.count()}개")
    cur.close()
    conn.close()
    return real_collection


# 우선 500개로 테스트
print("⏳ 실제 DB POI 인덱싱 시작 (500개 테스트)...")
real_collection = index_poi_from_db(limit=500)
```

```python
# 실제 DB 컬렉션으로 검색 테스트
def search_real_poi(query: str, top_k: int = 3) -> list[dict]:
    query_vec = embedder.encode(query, normalize_embeddings=True).tolist()
    results = real_collection.query(
        query_embeddings=[query_vec],
        n_results=top_k,
        include=["metadatas", "distances"]
    )
    return [
        {**meta, "similarity": round(1 - dist, 3)}
        for meta, dist in zip(results["metadatas"][0], results["distances"][0])
    ]

# 실제 DB 검색 테스트
results = search_real_poi("한옥 체험 전통 문화")
print("🔍 실제 DB 검색 결과: '한옥 체험 전통 문화'\n")
for r in results:
    print(f"  - {r['name']} ({r['sido']}) 유사도: {r['similarity']}")
    print(f"    {r['address']}")
```

지금까지 만든 것을 하나로 연결해서 최종 동작을 확인합니다.

```python
print("=" * 60)
print("K-Ride RAG 시스템 최종 점검")
print("=" * 60)

checks = {
    "Ollama 서버": False,
    f"LLM ({MODEL_NAME})": False,
    "임베딩 모델 (bge-m3)": False,
    "ChromaDB 샘플 (10개)": False,
    "RAG 검색": False,
}

# 1. Ollama 연결
try:
    ollama.list()
    checks["Ollama 서버"] = True
except:
    pass

# 2. LLM 모델
try:
    models = [m.get("name", m.get("model", "")) for m in ollama.list().get("models", [])]
    if MODEL_NAME in models:
        checks[f"LLM ({MODEL_NAME})"] = True
except:
    pass

# 3. 임베딩 모델
try:
    test_vec = embedder.encode("test")
    if len(test_vec) == 1024:
        checks["임베딩 모델 (bge-m3)"] = True
except:
    pass

# 4. ChromaDB
try:
    if collection.count() >= 10:
        checks["ChromaDB 샘플 (10개)"] = True
except:
    pass

# 5. RAG 검색
try:
    r = search_poi("서울 관광")
    if len(r) > 0:
        checks["RAG 검색"] = True
except:
    pass

for name, ok in checks.items():
    icon = "✅" if ok else "❌"
    print(f"  {icon} {name}")

all_ok = all(checks.values())
print()
if all_ok:
    print("🎉 모든 항목 정상! RAG 시스템 준비 완료입니다.")
else:
    failed = [k for k, v in checks.items() if not v]
    print(f"⚠️  다음 항목을 확인하세요: {', '.join(failed)}")
```
